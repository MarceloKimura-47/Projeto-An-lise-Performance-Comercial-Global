```python
# Criando lista com os caminhos dos arquivos

lista_caminhos_bronze = ['C:/Users/Kimura/Desktop/Projetos/Analise_Performance_Comercial_Global/1_bronze/devolucao.parquet',
                         'C:/Users/Kimura/Desktop/Projetos/Analise_Performance_Comercial_Global/1_bronze/vendas.parquet',
                         'C:/Users/Kimura/Desktop/Projetos/Analise_Performance_Comercial_Global/1_bronze/clientes.parquet',
                         'C:/Users/Kimura/Desktop/Projetos/Analise_Performance_Comercial_Global/1_bronze/localidades.parquet',
                         'C:/Users/Kimura/Desktop/Projetos/Analise_Performance_Comercial_Global/1_bronze/lojas.parquet',
                         'C:/Users/Kimura/Desktop/Projetos/Analise_Performance_Comercial_Global/1_bronze/produtos.parquet']
```

```python
# Criando os Dataframes com os arquivos.

df_devolucao = pd.read_parquet(lista_caminhos_bronze[0])
df_vendas = pd.read_parquet(lista_caminhos_bronze[1])
df_clientes = pd.read_parquet(lista_caminhos_bronze[2])
df_localidades = pd.read_parquet(lista_caminhos_bronze[3])
df_lojas = pd.read_parquet(lista_caminhos_bronze[4])
df_produtos = pd.read_parquet(lista_caminhos_bronze[5])
```

### Iremos inicar a OBT com os JOINs

Para isso iremos realizar vários LEFT JOIN com a tabela principal sendo a de Vendas (tb_fato).\
Iremos identificar quais colunas iremos usar para a conexão.

```python
df_vendas.info()
```

```python
df_clientes.info()
df_produtos.info()
df_devolucao.info()
```

### Iremos primeiro conectar ID CLIENTES e os SKUs

```python
join_01 = (
    df_vendas
    .merge(df_produtos, on='SKU', how='left')
    .merge(df_clientes, on='ID Cliente', how='left')
)
```

```python
join_01.info()
```

### Agora iremos conclui a conexão com o restante das tabelas

```python
df_localidades.info()
df_lojas.info()
```

```python
join_02 = (
    df_lojas
    .merge(df_localidades, left_on='id Localidade', right_on='ID Localidade' , how='left')
        )
join_02.head()
```

```python
df_OBT = (join_01
          .merge(join_02, on="ID Loja", how='left')
         )
```

```python
df_OBT.info()
df_OBT.head()
```

### Últimos ajustes:

Iremos agora realizar a limpeza de alguns dados redundantes no DataFrame.

Além disso iremos padronizar os nomes das colunas.


```python
# Excluindo as colunas de ingestão que importamos duplicadas.

df_OBT = df_OBT.drop(columns=[
    'data_ingestao_x_x', 'origem_arquivo_x_x',
    'data_ingestao_y_x', 'origem_arquivo_y_x',
    'data_ingestao_x_y', 'origem_arquivo_x_y',
    'data_ingestao_y_y', 'origem_arquivo_y_y',
    'id Localidade'])
```

```python
df_OBT.info()
df_OBT.head()
```

```python
# Padronizando os nomes das colunas do Dataframe

df_OBT.columns = (
    df_OBT.columns
    .str.lower()
    .str.strip()
    .str.replace(' ', '_')
    .str.normalize('NFKD')  # remove acentos
    .str.encode('ascii', errors='ignore')
    .str.decode('utf-8')
)
```

```python
df_OBT.info()
```

```python
df_OBT.to_parquet('C:/Users/Kimura/Desktop/Projetos/Analise_Performance_Comercial_Global/2_silver/One_Big_Table.parquet')
```

### OBT de Devolução

Mesmo processo mas para a tabela de devolução.

```python
join_03 = (
        df_devolucao
        .merge(df_produtos, on='SKU', how='left')
        .merge(join_02, on='ID Loja', how='left')
         )
```

```python
join_03 = join_03.drop(columns=[
    'data_ingestao_x_x', 'origem_arquivo_x_x',
    'data_ingestao_y_x', 'origem_arquivo_y_x',
    'data_ingestao_x_y', 'origem_arquivo_x_y',
    'data_ingestao_y_y', 'origem_arquivo_y_y',
    'id Localidade'])
```

```python
join_03.columns = (
    join_03.columns
    .str.lower()
    .str.strip()
    .str.replace(' ', '_')
    .str.normalize('NFKD')  # remove acentos
    .str.encode('ascii', errors='ignore')
    .str.decode('utf-8')
)
```

```python
df_OBT_02 = join_03
df_OBT_02.to_parquet('C:/Users/Kimura/Desktop/Projetos/Analise_Performance_Comercial_Global/2_silver/One_Big_Table_02.parquet')
```
