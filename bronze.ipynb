```python
import pandas as pd
from datetime import  datetime
```

## Importação dos arquivos da camada RAW

Importei os arquivos da camanada raw em um dicionário com a chave sendo o nome do dataframe e o valor sendo o caminho do arquivos dentro da camada raw.\
Isso irá facilitar no momento em que iremos criar as colunas com as informações de `data de ingestão` e `caminho arquivos`.

```python
# Criação da variável que vai conter o dicionário com o nome do dataframe + caminho arquivo.

arquivos_raw = {
    "df_devolucao": 'C:/Users/Kimura/Desktop/Projetos/Analise_Performance_Comercial_Global/0_raw/devolucao.csv',
    "df_vendas": 'C:/Users/Kimura/Desktop/Projetos/Analise_Performance_Comercial_Global/0_raw/vendas_2020.csv',
    "df_clientes": 'C:/Users/Kimura/Desktop/Projetos/Analise_Performance_Comercial_Global/0_raw/clientes.csv',
    "df_localidades": 'C:/Users/Kimura/Desktop/Projetos/Analise_Performance_Comercial_Global/0_raw/localidades.csv',
    "df_lojas": 'C:/Users/Kimura/Desktop/Projetos/Analise_Performance_Comercial_Global/0_raw/lojas.csv',
    "df_produtos": 'C:/Users/Kimura/Desktop/Projetos/Analise_Performance_Comercial_Global/0_raw/produtos.csv'
            }
```

```python
# Aqui nós criamos um loop for para criar os Dataframes e adicionar em cada um deles a informação de Data de ingestão e Origem dos arquivos.

dfs = {}

for nome_df, caminho in arquivos_raw.items():
    df = pd.read_csv(caminho, delimiter=';')
    df["data_ingestao"] = datetime.now().strftime("%Y-%m-%d %H:%M")
    df["origem_arquivo"] = caminho  
    dfs[nome_df] = df
```

```python
# Atribuição dos dataframes.

df_devolucao = dfs["df_devolucao"]
df_vendas = dfs["df_vendas"]
df_clientes = dfs["df_clientes"]
df_localidades = dfs["df_localidades"]
df_lojas = dfs["df_lojas"]
df_produtos = dfs["df_produtos"]
```

## Checagem nos arquivos:
Iremos agora realizar algumas checagens nos dataframes E realizar mais alguns ajustes se necessários.

```python
# Precisamos alterar o tipo de dado da data
# Convertendo o tipo de dado das colunas de datas.
df_devolucao['Data Devolução'] = pd.to_datetime(df_devolucao['Data Devolução'])
df_devolucao['data_ingestao'] = pd.to_datetime(df_devolucao['data_ingestao'])
df_devolucao.info()
df_devolucao.head()
```

```python
df_vendas['Data da Venda'] = pd.to_datetime(df_vendas['Data da Venda'])
df_vendas['data_ingestao'] = pd.to_datetime(df_vendas['data_ingestao'])
df_vendas.info()
df_vendas.head()
```

```python
# Eliminando colunas não necessárias.

df_clientes = df_clientes.drop('Unnamed: 10', axis=1)
df_clientes = df_clientes.drop('Unnamed: 11', axis=1)
df_clientes.info()
df_clientes.head()
```

```python
# Remover a parte da hora, se presente
df_clientes['Data Nascimento'] = df_clientes['Data Nascimento'].astype(str).str.split(' ').str[0]

# Agora converta as datas para o formato correto
df_clientes['Data Nascimento'] = pd.to_datetime(df_clientes['Data Nascimento'], format='%m/%d/%Y', errors='coerce')

df_clientes['data_ingestao'] = pd.to_datetime(df_clientes['data_ingestao'])
```

```python
# OK

df_localidades.info()
df_localidades.head()
```

```python
# Nesse caso a coluna de DOCUMENTO iremos converter para String

df_lojas.info()
df_lojas.head()

df_lojas['Documento Gerente'] = df_lojas['Documento Gerente'].astype(str)
```

```python
# Excluindo a coluna "OBS" (tem apenas 4 registros)

df_produtos = df_produtos.drop('OBS',axis=1)
df_produtos.info()
df_produtos.head()
```
